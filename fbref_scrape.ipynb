Python script to scrape player in-game statistics data from fbref.com. Extracted data from 17 professional soccer leagues around the world.

```python
# libraries
import pandas as pd
import numpy as np
import requests
import time
import random

years = list(np.arange(2018,2023,1))
years_special = ['2018-2019','2019-2020','2020-2021','2021-2022','2022-2023']
numbers = [22,24,21,9,12,20,11,13,32,31,23,37,10,17,33,60,18]
stats = ['passing','defense','shooting']
leagues = ['-Major-League-Soccer-Stats','-Serie-A-Stats','-Primera-Division-Stats','-Premier-League-Stats',
           '-La-Liga-Stats','-Bundesliga-Stats','-Serie-A-Stats','-Ligue-1-Stats','-Primeira-Liga-Stats',
          '-Liga-MX-Stats','-Eredivisie-Stats','-Belgian-Pro-League-Stats','-Championship-Stats',
          '-Segunda-Division-Stats','-2-Bundesliga-Stats','-Ligue-2-Stats','-Serie-B-Stats']
league_names = ['U.S.A. (MLS)', 'Brazil (First Division)', 'Argentina (Premier Division)', 'England (Premier Division)',
                'Spain (First Division)','Germany (Bundesliga)','Italy (Serie A)','France (Ligue 1 Uber Eats)',
                'Portugal (Premier League)', 'Mexico (First Division)', 'Holland (Eredivisie)','Belgium (Jupiler Pro League)', 'England (Sky Bet Championship)',
                'Spain (Second Division)','Germany (Bundesliga 2)','France (Ligue 2 BKT)','Italy (Serie B)']

defense_list = []
passing_list = []
shooting_list = []
count = 0
for n in range(len(numbers)):
    if count < 3:
        for s in stats:
            for y in years:
                data = 'https://fbref.com/en/comps/' + str(numbers[count]) + '/' + str(y) + '/' + s + '/' + str(y) + leagues[count]
                response = requests.get(data).text.replace('<!--', '').replace('-->', '')
                df = pd.read_html(response, header=1)[2]
                df['Based'] = league_names[count]
                if s == 'passing':
                    passing_list.append(df)
                elif s == 'defense':
                    defense_list.append(df)
                else:
                    shooting_list.append(df)
                time.sleep(3)
                print(count)
    else:
        for t in stats:
            for z in years_special:
                data = 'https://fbref.com/en/comps/' + str(numbers[count]) + '/' + z + '/' + t + '/' + z + leagues[count]
                response = requests.get(data).text.replace('<!--', '').replace('-->', '')
                df = pd.read_html(response, header=1)[2]
                df['Based'] = league_names[count]
                if t == 'passing':
                    passing_list.append(df)
                elif t == 'defense':
                    defense_list.append(df)
                else:
                    shooting_list.append(df)
                time.sleep(3)
                print(count)

    count += 1
```

```python
shot = []
defense = []
passing = []
for i in range(len(years_special)):
    shot.append(shooting_list[i::len(years_special)])
    defense.append(defense_list[i::len(years_special)])
    passing.append(passing_list[i::len(years_special)])
```

```python
final_datasets = []
for y in range(len(years_special)):
    for k in range(3):
        league_names = ['U.S.A. (MLS)', 'Brazil (First Division)', 'Argentina (Premier Division)', 'England (Premier Division)',
                'Spain (First Division)','Germany (Bundesliga)','Italy (Serie A)','France (Ligue 1 Uber Eats)',
                'Portugal (Premier League)', 'Mexico (First Division)', 'Holland (Eredivisie)','Belgium (Jupiler Pro League)', 'England (Sky Bet Championship)',
                'Spain (Second Division)','Germany (Bundesliga 2)','France (Ligue 2 BKT)','Italy (Serie B)']
        if stats[k] == 'passing':
            count = 0
            for df in passing[y]:
                # rename columns
                df = df.fillna(0)

                df = df.drop(columns=['Rk', 'Nation', 'Pos', 'A-xAG',
                                     '1/3', 'PPA', 'CrsPA', 'Matches','TotDist', 'PrgDist'])

                df.rename(columns={'Cmp':'Total Cmp', 'Att': 'Total Att', 'Cmp%':'Total Cmp%','Cmp.1':'Short Cmp', 
                                   'Att.1':'Short Att','Cmp%.1':'Short Cmp%', 'Cmp.2':'Medium Cmp', 'Att.2':'Medium Att',
                                   'Cmp%.2':'Medium Cmp%', 'Cmp.3':'Long Cmp', 'Att.3':'Long Att', 'Cmp%.3':'Long Cmp%'}, inplace = True)
                passing[y][count] = df

                count += 1

        elif stats[k] == 'defense':
            count = 0
            for df in defense[y]:

                # rename columns
                df = df.fillna(0)

                df['Age'] = df['Age'].str[:2]
                df['Position_2'] = df['Pos'].str[3:]
                df['Position'] = df['Pos'].str[:2]
                df['Nation'] = df['Nation'].str.split(' ').str.get(1)

                df['Position'] = df['Position'].replace({'MF': 'Midfielder', 'DF': 'Defender', 'FW': 'Forward', 'GK': 'Goalkeeper'})
                df['Position_2'] = df['Position_2'].replace({'MF': 'Midfielder', 'DF': 'Defender',
                                                                 'FW': 'Forward', 'GK': 'Goalkeeper'})

                df = df.drop(columns=['Rk', 'Pos','Matches','Nation','Position_2','Tkl+Int',
                                      'Def 3rd', 'Mid 3rd', 'Att 3rd','Tkl.1','Att', 'Lost','Blocks','Position'])

                df.rename(columns={'Sh':'Blocks Sh','Pass':'Blocks Pass'}, inplace = True)
                defense[y][count] = df

                count += 1

        else:
            count = 0
            for df in shot[y]:

                # rename columns
                df = df.fillna(0)

                df['Age'] = df['Age'].str[:2]
                df['Position_2'] = df['Pos'].str[3:]
                df['Position'] = df['Pos'].str[:2]
                df['Nation'] = df['Nation'].str.split(' ').str.get(1)

                df['Position'] = df['Position'].replace({'MF': 'Midfielder', 'DF': 'Defender', 'FW': 'Forward', 'GK': 'Goalkeeper'})
                df['Position_2'] = df['Position_2'].replace({'MF': 'Midfielder', 'DF': 'Defender',
                                                                 'FW': 'Forward', 'GK': 'Goalkeeper'})

                df = df.drop(columns=['Rk', 'Pos','Matches','Nation','Position_2', 'Dist'])

                shot[y][count] = df

                count += 1

    df_shot = pd.concat(shot[y])
    df_pass = pd.concat(passing[y])
    df_def = pd.concat(defense[y])
    
    final = pd.merge(pd.merge(df_shot,df_def,on=['Player','Squad','Born','Based','Age','90s']),df_pass,on=['Player','Squad','Born','Based','Age','90s'])
    
    final_datasets.append(final)
```
