```python
import pandas as pd

players_2020 = pd.read_csv('2020.csv')
players_2021 = pd.read_csv('2021.csv')
players_2022 = pd.read_csv('2022.csv')
fm = pd.read_csv('players.csv')
```

```python
fm = fm.drop(['Unnamed: 0','DoB'], axis=1).dropna()
```

```python
player_stats_2022 = pd.merge(fm,players_2022,on=['Player','Based','Born'])

#begin fixing missing/incorrect data
#drop duplicate players
player_stats_2022 = player_stats_2022.drop_duplicates('UID')

#remove players with first letter of squad and club columns not the same
dup = player_stats_2022[player_stats_2022.duplicated(['Player'], keep=False)]
dup_players = dup['Player'].tolist()
for i in dup_players:
    player_stats_2022 = player_stats_2022.loc[~((player_stats_2022['Player'] == i) & (player_stats_2022.Club.str[0]!=player_stats_2022.Squad.str[0]))]

# #drop special cases
final_players_2022 = player_stats_2022.drop_duplicates( 
  subset = ['Player','Born','Squad','Club'], 
  keep = 'last').reset_index(drop = True) 

# #final player stat dataset with correct unique player data
final_players_2022
```

```python
phys_stats = final_players_2022[['UID', 'Player', 'Nat', 'Preferred Foot',
       'Height (cm)', 'Weight (kg)', 'birth_month', 'Born', 'Club']]
player_stats_2021 = pd.merge(phys_stats,players_2021,on=['Player','Born'])
dups = player_stats_2021[player_stats_2021.duplicated(['UID'], keep=False)]
dups[['Player','Club','Squad']]

#remove duplicate players
dup = player_stats_2021[player_stats_2021.duplicated(['UID'], keep=False)]

dup['Player']
dup_players = dup['Player'].tolist()
for i in dup_players:
    player_stats_2021 = player_stats_2021.loc[~((player_stats_2021['Player'] == i) & (player_stats_2021.Club.str[0]!=player_stats_2021.Squad.str[0]))]
    dup = dup.loc[~((dup['Player'] == i) & (dup.Club.str[0]!=dup.Squad.str[0]))]

dup2 = dup[dup.duplicated(['UID'], keep=False)]
dup_players2 = dup2['Player'].tolist()
for i in dup_players2:
    final_players_2021 = player_stats_2021.loc[~((player_stats_2021['Player'] == i) & (player_stats_2021.Club.str[0:2]!=player_stats_2021.Squad.str[0:2]))]
    
#drop any remaining duplicate players
final_players_2021 = final_players_2021.drop_duplicates('UID').drop(['Unnamed: 0','Total Cmp','Total Att', 
                                                                     'Total Cmp%','Short Att','Medium Att',
                                                                     'Long Att','Tkl', 'Sh','Squad','SoT','Based'], axis=1)
final_players_2021.columns
```

```python
phys_stats = final_players_2021[['UID', 'Player', 'Nat', 'Preferred Foot',
       'Height (cm)', 'Weight (kg)', 'birth_month', 'Born', 'Club']]
player_stats_2020 = pd.merge(phys_stats,players_2020,on=['Player','Born']).sort_values(['Player', '90s'], ascending=[True, False])
season_2020 = player_stats_2020.drop_duplicates(['UID']).drop(['Total Cmp','Total Att', 'Total Cmp%','Short Att','Medium Att',
                                                               'Long Att','Tkl', 'Sh','Squad','SoT','Based'],axis=1)

season_2020

#'xG', 'npxG', 'npxG/Sh','G-xG', 'np:G-xG','xAG', 'xA','Age',
```

```python
season_2021 = final_players_2021[final_players_2021['UID'].isin(season_2020['UID'])].sort_values(['Player', '90s'], ascending=[True, False])
season_2021
```

```python
cols = season_2021.columns.to_list()

season_2022 = final_players_2022[final_players_2022['UID'].isin(season_2020['UID'])].sort_values(['Player', '90s'], ascending=[True, False])
season_2022 = season_2022.drop(['Total Cmp','Total Att', 'Total Cmp%','Short Att','Medium Att',
                                'Long Att','Tkl', 'Sh','Squad','SoT','Based'],axis=1)
season_2022 = season_2022[cols]
season_2022

#'xG','npxG', 'npxG/Sh', 'G-xG', 'np:G-xG','xAG', 'Age','xA','Squad'
```

```python
cols = ['Gls', 'FK', 'PK', 'PKatt', 'xG', 'npxG',
        'G-xG', 'np:G-xG', 'TklW','Blocks Sh', 'Blocks Pass',
        'Int', 'Clr', 'Err', 'Short Cmp',
        'Medium Cmp', 'Long Cmp', 'Ast', 'xAG', 'xA', 'KP', 'PrgP']

season_2020[cols] = season_2020[cols].div(season_2020['90s'],axis=0).round(2)
season_2020.rename(columns = {'Gls':'Gls/90', 'FK':'FK/90', 'PK':'PK/90', 'PKatt':'PKatt/90', 'xG':'xG/90',
                        'npxG':'npxG/90','G-xG':'G-xG/90', 'np:G-xG':'np:G-xG/90', 'TklW':'TklW/90',
                        'Blocks Sh':'Blocks Sh/90', 'Blocks Pass':'Blocks Pass/90','Int':'Int/90', 
                        'Clr':'Clr/90', 'Err':'Err/90', 'Short Cmp':'Short Cmp/90','Medium Cmp':'Medium Cmp/90',
                        'Long Cmp':'Long Cmp/90', 'Ast':'Ast/90', 'xAG':'xAG/90', 'xA':'xA/90', 'KP':'KP/90',
                        'PrgP':'PrgP/90'}, inplace = True)

season_2021[cols] = season_2021[cols].div(season_2021['90s'],axis=0).round(2)
season_2021.rename(columns = {'Gls':'Gls/90', 'FK':'FK/90', 'PK':'PK/90', 'PKatt':'PKatt/90', 'xG':'xG/90',
                        'npxG':'npxG/90','G-xG':'G-xG/90', 'np:G-xG':'np:G-xG/90', 'TklW':'TklW/90',
                        'Blocks Sh':'Blocks Sh/90', 'Blocks Pass':'Blocks Pass/90','Int':'Int/90', 
                        'Clr':'Clr/90', 'Err':'Err/90', 'Short Cmp':'Short Cmp/90','Medium Cmp':'Medium Cmp/90',
                        'Long Cmp':'Long Cmp/90', 'Ast':'Ast/90', 'xAG':'xAG/90', 'xA':'xA/90', 'KP':'KP/90',
                        'PrgP':'PrgP/90'}, inplace = True)

season_2022[cols] = season_2022[cols].div(season_2022['90s'],axis=0).round(2)
season_2022.rename(columns = {'Gls':'Gls/90', 'FK':'FK/90', 'PK':'PK/90', 'PKatt':'PKatt/90', 'xG':'xG/90',
                        'npxG':'npxG/90','G-xG':'G-xG/90', 'np:G-xG':'np:G-xG/90', 'TklW':'TklW/90',
                        'Blocks Sh':'Blocks Sh/90', 'Blocks Pass':'Blocks Pass/90','Int':'Int/90', 
                        'Clr':'Clr/90', 'Err':'Err/90', 'Short Cmp':'Short Cmp/90','Medium Cmp':'Medium Cmp/90',
                        'Long Cmp':'Long Cmp/90', 'Ast':'Ast/90', 'xAG':'xAG/90', 'xA':'xA/90', 'KP':'KP/90',
                        'PrgP':'PrgP/90'}, inplace = True)

season_2022 = season_2022[season_2022['90s']>=1.0]
season_2021 = season_2021[season_2021['90s']>=1.0]
season_2020 = season_2020[season_2020['90s']>=1.0]
```

```python
stats_20_21 = season_2020.merge(season_2021, on=['UID', 'Player', 'Nat', 'Preferred Foot',
       'Height (cm)', 'Weight (kg)', 'birth_month', 'Born', 'Club','Position'])
stats_20_21 = stats_20_21.drop(['Unnamed: 0'], axis=1)

stats_21_22 = season_2021.merge(season_2022, on=['UID', 'Player', 'Nat', 'Preferred Foot',
       'Height (cm)', 'Weight (kg)', 'birth_month', 'Born', 'Club','Position'])
#stats_21_22 = stats_21_22.drop(['Club'], axis=1)

stats_21_22 = stats_21_22[stats_21_22['UID'].isin(stats_20_21['UID'])].reset_index(drop=True)
stats_20_21 = stats_20_21[stats_20_21['UID'].isin(stats_21_22['UID'])].reset_index(drop=True)

stats_20_21
```
